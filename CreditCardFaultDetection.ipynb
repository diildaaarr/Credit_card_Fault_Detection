```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head(5)
```

```python
credit_card_data.tail(5)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')  # Specify the imputation strategy
credit_card_data.iloc[:, :] = imputer.fit_transform(credit_card_data)
```

```python
credit_card_data.isnull().sum()
```

```python
# Separate legitimate and fraudulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

choose random 492 legit

```python
new_data1 = legit.sample(n=492)
new_data2 = fraud
new_data = pd.concat([new_data1, new_data2])
```

```python
new_data.sample(5)
```

```python
X = new_data.drop(columns='Class', axis=1)
Y = new_data['Class']

# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform it
X_train_scaled = scaler.fit_transform(X_train)

# Transform the test data using the same scaler
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

```python
new_data['Class'].value_counts()
```

```python
# This LogisticRegression because we need is the
model=LogisticRegression(max_iter=1000)

model.fit(X_train_scaled,Y_train)
Y_pred=model.predict(X_test_scaled)
```

```python
accuracy = accuracy_score(Y_test,Y_pred)
accuracy
```

```python
cv_scores = cross_val_score(model, X_train_scaled, Y_train, cv=5)
cv_scores
```

```python
print(f"the cv score mean is as {cv_scores.mean()}")
```
